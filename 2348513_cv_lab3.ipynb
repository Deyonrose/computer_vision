image enhancement techniques using Built-in function



```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load image
img = cv2.imread('/content/azalea.jpg', 0)

# Built-in Functions

# 1. Image Negative
def built_in_image_negative(image):
    return 255 - image

# 2. Gray Level Slicing
def built_in_gray_level_slicing(image, lower_thresh, upper_thresh):
    result = np.copy(image)
    result[(image >= lower_thresh) & (image <= upper_thresh)] = 255
    return result

# 3. Histogram Equalization
def built_in_histogram_equalization(image):
    equ = cv2.equalizeHist(image)
    return equ
```

Printing the orginal Image

```python
# Display original image
plt.imshow(img, cmap='gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
```

printing using Built in functions

```python
# Apply built-in functions
plt.figure(figsize=(12,8))
# Image Negative
plt.subplot(1, 3, 1), plt.imshow(built_in_image_negative(img), cmap='gray')
plt.title('Built-in Image Negative'), plt.xticks([]), plt.yticks([])

# Gray Level Slicing
plt.subplot(1, 3, 2), plt.imshow(built_in_gray_level_slicing(img, 100, 200), cmap='gray')
plt.title('Built-in Gray Level Slicing'), plt.xticks([]), plt.yticks([])

# Histogram Equalization
plt.subplot(1, 3, 3), plt.imshow(built_in_histogram_equalization(img), cmap='gray')
plt.title('Built-in Histogram Equalization'), plt.xticks([]), plt.yticks([])
```

**User Defined functions**

```python
# User Defined Functions

# 1. Image Negative
def user_defined_image_negative(image):
    return 255 - image

# 2. Gray Level Slicing
def user_defined_gray_level_slicing(image, lower_thresh, upper_thresh):
    result = np.copy(image)
    result[(image >= lower_thresh) & (image <= upper_thresh)] = 255
    return result

# 3. Histogram Equalization
def user_defined_histogram_equalization(image):
    hist, bins = np.histogram(image.flatten(), 256, [0, 256])
    cdf = hist.cumsum()
    cdf_normalized = cdf * hist.max() / cdf.max()

    cdf_m = np.ma.masked_equal(cdf, 0)
    cdf_m = (cdf_m - cdf_m.min()) * 255 / (cdf_m.max() - cdf_m.min())
    cdf = np.ma.filled(cdf_m, 0).astype('uint8')

    result = cdf[image]
    return result
```

Printing user Defined Functions

```python
# Apply user-defined functions
plt.figure(figsize=(12,8))
# Image Negative
plt.subplot(1, 3, 1), plt.imshow(user_defined_image_negative(img), cmap='gray')
plt.title('User-defined Image Negative'), plt.xticks([]), plt.yticks([])

# Gray Level Slicing
plt.subplot(1, 3, 2), plt.imshow(user_defined_gray_level_slicing(img, 100, 200), cmap='gray')
plt.title('User-defined Gray Level Slicing'), plt.xticks([]), plt.yticks([])

#Histogram Equalization
plt.subplot(1, 3, 3), plt.imshow(user_defined_histogram_equalization(img), cmap='gray')
plt.title('User-defined Histogram Equalization'), plt.xticks([]), plt.yticks([])
```
